## Check feasibility with CVX

```python
n, m = 100, 100
```

### 1

```python
def ex_1_create(m, n, seed=42):
    np.random.seed(seed)
    random.seed(seed)

    A = np.random.randn(m, n)
    b = np.random.randn(m)
    
    return A, b
```

### 2. 

```python
def ex_2_create(m, n, seed=42):
    np.random.seed(seed)
    random.seed(seed)

    A = np.random.randn(m, n)

    x = np.abs(np.random.randn(n))

    b = A @ x
    
    return A, b
```

```python
A, b = ex_2_create(m, n)

x = cp.Variable(n)
prob = cp.Problem(cp.Minimize(0), [A @ x == b, x >= 0])
result = prob.solve()

if x.value is not None:
    print("Exists")
else:
    print("Infeasible")
```

### 3(a)

```python
def ex_3a_create(m, n, seed=42):
    np.random.seed(seed)
    random.seed(seed)

    A = np.random.randn(m, n)

    x = np.random.randn(n)

    e = np.abs(np.random.randn(m))

    b = A @ x + e
    
    return A, b
```

```python
A, b = ex_3a_create(m, n)

x = cp.Variable(n)
prob = cp.Problem(cp.Minimize(0), [A @ x <= b])
result = prob.solve()

if x.value is not None:
    print("Exists")
else:
    print("Infeasible")
```

### 3(b)

```python
def ex_3b_create(m, n, seed=42):
    np.random.seed(seed)
    random.seed(seed)

    A = np.random.randn(m, n)
    
    v1 = np.random.randn(n)
    
    e = (3 + np.random.rand(n))# + np.abs(np.random.randn(n))
    v2 = v1 + e
    
    e = (1 + np.random.rand(n)) + np.minimum(1, np.abs(np.random.randn(n)))
    x = v1 + e

    b = A @ x
    
    return A, b, v1, v2
```

```python
A, b, v1, v2 = ex_3b_create(m, n)

x = cp.Variable(n)
prob = cp.Problem(cp.Minimize(0), [A @ x == b, v1 <= x, x <= v2])
result = prob.solve()

if x.value is not None:
    print("Exists")
else:
    print("Infeasible")
```

### 3(c)

```python
def ex_3c_create(m, n, seed=42):
    np.random.seed(seed)
    random.seed(seed)

    A = np.random.randn(m, n)
    
    v1 = np.random.randn(n)
    
    e = (3 + np.random.rand(n))# + np.abs(np.random.randn(n))
    v2 = v1 + e
    
    e = (1 + np.random.rand(n)) + np.minimum(1, np.abs(np.random.randn(n)))
    x = v1 + e

    e = np.abs(np.random.randn(m))
    b = A @ x + e
    
    return A, b, v1, v2
```

```python
A, b, v1, v2 = ex_3c_create(m, n)

x = cp.Variable(n)
prob = cp.Problem(cp.Minimize(0), [A @ x <= b, v1 <= x, x <= v2])
result = prob.solve()

if x.value is not None:
    print("Exists")
else:
    print("Infeasible")
```

## Projections

```python
def proj_hyperplane(x, a, b):
    return x - ( (x @ a - b) / (a @ a) ) * a
```

```python
def proj_halfspace(x, a, b):
    return x + ( max(0, b - a @ x) / (a @ a) ) * a
```

```python
def proj_box(x, v1, v2):
    return np.minimum( np.maximum(x, v1), v2 )
```

```python
def proj_nonnegative(x):
    return np.maximum(x, 0)
```

```python
def distance(x, proj_func):
    return np.linalg.norm(x - proj_func(x))
```

## Algos

```python
def cyclic_algorithm(projection_functions, n, max_k):
    
    np.random.seed(43)
    random.seed(43)
    x = np.random.randn(n)
    
    distances = []
    
    for _ in range(max_k):
        
        for proj_func in projection_functions:
            x = proj_func(x)
        
        max_dist = max([
            distance(x, proj_func)
            for proj_func in projection_functions
        ])
        distances.append(max_dist)
        
    return distances
```

```python
def greedy_algorithm(projection_functions, n, max_k):
    
    np.random.seed(43)
    random.seed(43)
    x = np.random.randn(n)
    
    distances = []
    
    for _ in range(max_k):
        
        i_of_max_dist = int(np.argmax([
            distance(x, proj_func)
            for proj_func in projection_functions
        ]))
        
        proj_func = projection_functions[i_of_max_dist]
        
        x = proj_func(x)
        
        max_dist = max([
            distance(x, proj_func)
            for proj_func in projection_functions
        ])
        distances.append(max_dist)
        
    return distances
```

```python
from functools import partial
```

## Ex. 1

```python
for m, n, max_k in [(200, 200, 2000), (130, 200, 1000), (200, 130, 1000)]:

    A, b = ex_1_create(m, n)

    proj_hyper = partial(proj_hyperplane, a=np.ones(n), b=1)

    funcs = [proj_hyper, proj_nonnegative]

    diffs_cyclic = cyclic_algorithm(funcs, n, max_k)
    diffs_greedy = greedy_algorithm(funcs, n, max_k)


    pdf = PdfPages(f'plots/2.10_ex_1_{m}_{n}.pdf')
    plt.figure(figsize=(12, 3))

    markevery=int(max_k/10)
    plt.plot(list(range(len(diffs_cyclic))), diffs_cyclic, label=f'Cyclic', marker='^', markevery=markevery, ms=9)
    plt.plot(list(range(len(diffs_greedy))), diffs_greedy, label=f'Greedy', marker='x', markevery=markevery, ms=9)

    # Add grid
    plt.grid(True, linestyle='--', alpha=0.6)

    plt.legend(ncols=2)

    # Label the axes
    plt.xlabel('$k$')
    plt.ylabel('max($d_{C_i}(x^k)$)')

    #plt.xscale('log')
    plt.yscale('log')

    plt.title(f"$S_1 = _n$ , m = {m} , n = {n}")

    plt.subplots_adjust(bottom=0.22, top=0.85)

    pdf.savefig(plt.gcf()) # Save the current figure
    plt.show()
    plt.close()
    pdf.close()
```

```python
for m, n, max_k in [(200, 200, 5000), (130, 200, 600), (200, 130, 600)]:

    A, b = ex_2_create(m, n)

    funcs = []

    for i in range(m):
        funcs.append(
            partial(proj_hyperplane, a=A[i,:], b=b[i])
        )

    funcs.append(
        proj_nonnegative
    )


    diffs_cyclic = cyclic_algorithm(funcs, n, max_k)
    diffs_greedy = greedy_algorithm(funcs, n, max_k)


    pdf = PdfPages(f'plots/2.10_ex_2_{m}_{n}.pdf')
    plt.figure(figsize=(12, 3))

    markevery=int(max_k/10)
    plt.plot(list(range(len(diffs_cyclic))), diffs_cyclic, label=f'Cyclic', marker='^', markevery=markevery, ms=9)
    plt.plot(list(range(len(diffs_greedy))), diffs_greedy, label=f'Greedy', marker='x', markevery=markevery, ms=9)

    # Add grid
    plt.grid(True, linestyle='--', alpha=0.6)

    plt.legend()

    # Label the axes
    plt.xlabel('$k$')
    plt.ylabel('max($d_{C_i}(x^k)$)')

    #plt.xscale('log')
    plt.yscale('log')

    plt.title("$S = ${$x : Ax = b, x  0 $} , "+f"m = {m} , n = {n}")

    plt.subplots_adjust(bottom=0.22, top=0.85)

    pdf.savefig(plt.gcf()) # Save the current figure
    plt.show()
    plt.close()
    pdf.close()
```
